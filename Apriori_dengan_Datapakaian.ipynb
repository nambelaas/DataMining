```python
import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend import frequent_patterns
from mlxtend.frequent_patterns import apriori
```

```python
#Membaca File
```

```python
dp = pd.read_csv('DataPakaian.csv')
dp.head(11)
```

```python
datas = dp.drop(["Transaksi"], axis = 1)
datas.head(11)
```

```python
df1 = frequent_patterns.apriori(datas, min_support=0.04,use_colnames=True)
df1['length'] = df1['itemsets'].apply(lambda x:len(x))
df1.sort_values(by=['support','length'],ascending=False)
```

```python
association_rules = frequent_patterns.association_rules(df1, metric = "confidence", min_threshold = 0.8)
association_rules
```
