```python
import numpy as np
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend import frequent_patterns
from mlxtend.frequent_patterns import apriori
```

```python
#Membaca File
```

```python
dp = pd.read_csv('DataPakaian2.csv',names=['Produk'])
dp
```

```python
dp.columns
```

```python
dp.values
```

```python
data = list(dp["Produk"].apply(lambda x:x.split(',')))
data
```

```python
txn_encode = TransactionEncoder()
predictions = txn_encode.fit(data).transform(data)
features = pd.DataFrame(predictions,columns=txn_encode.columns_)
features = features.replace(False,0)
features
```

```python
df1 = frequent_patterns.apriori(features, min_support=0.04,use_colnames=True)
df1['length'] = df1['itemsets'].apply(lambda x:len(x))
df1.sort_values(by=['support','length'],ascending=False)
```

```python
association_rules = frequent_patterns.association_rules(df1, metric = "confidence", min_threshold = 0.8)
association_rules
```
